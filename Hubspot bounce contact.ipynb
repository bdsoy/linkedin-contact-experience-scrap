```python
import requests, time, random
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
import pandas as pd
namelist=[]
df=pd.read_excel('hubspot.xlsx')
for i in df.Recipient:
    namelist.append(i.split('@')[0]+' '+i.split('@')[1].split('.')[0])
```

```python
browser = webdriver.Chrome('C:/Users/Liting Cai/PycharmProjects/selenium/chromedriver')
browser.get('https://www.linkedin.com/login')
username = ''
password = ''

elementID = browser.find_element_by_id('username')
elementID.send_keys(username)

elementID = browser.find_element_by_id('password')
elementID.send_keys(password)

elementID.submit()
```

```python
def search_contact_by_name(name):
    import time
    time.sleep(5)
    browser.find_element_by_class_name('search-global-typeahead__input').clear()
    elementID = browser.find_element_by_class_name('search-global-typeahead__input')
    elementID.send_keys(name)
    elementID.send_keys(Keys.ENTER)
    import time
    time.sleep(5)
    try:
        browser.find_element_by_class_name('name').click()
        import time
        time.sleep(5)
        SCROLL_PAUSE_TIME = 5
        last_height = browser.execute_script("return document.body.scrollHeight")
        
        for i in range(3):
            browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            import time
            time.sleep(SCROLL_PAUSE_TIME)
            new_height = browser.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                break
            last_height = new_height
        
        src = browser.page_source
        soup = BeautifulSoup(src, 'lxml')
#         containers=soup.findAll("ul",{"class":"pv-profile-section__section-info"})
#         return containers
        return soup

    except:
        try:
            import time
            time.sleep(5)
            browser.find_element_by_class_name('search-global-typeahead__input').clear()
            elementID = browser.find_element_by_class_name('search-global-typeahead__input')
            elementID.send_keys(name.split(' ')[0])
            elementID.send_keys(Keys.ENTER)
            import time
            time.sleep(5)
            browser.find_element_by_class_name('name').click()
            import time
            time.sleep(5)
            src = browser.page_source
            soup = BeautifulSoup(src, 'lxml')
            return soup

        except:
            print('this pp {} not in linkedin'.format(name.split(' ')[0]))
            return []
```

```python
henkilo_info=[]
henkilo_job=[]
for i in namelist:
    soup=search_contact_by_name(i)
    if soup != []:
        name_loc=soup.find('div',{'class','flex-1 mr5'}).find_all('ul')
        name=name_loc[0].find('li').get_text()
        loc=name_loc[1].find('li').get_text().strip()
        profile_title=soup.find('div',{'class','flex-1 mr5'}).find('h2').get_text().strip()
        henkilo_info.append([name, loc, profile_title])


        if soup.find('section',{'id':'experience-section'})!=None:
            exp_section=soup.find('section',{'id':'experience-section'}).find('ul')
            job_title=exp_section.find('div').find('a').find('h3').get_text().strip()
            work_period=exp_section.find('div').find('a').findAll('h4')[0].find_all('span')[1].get_text().strip()
            if exp_section.find('div').find('a').findAll('p')!= []:
                company_name=exp_section.find('div').find('a').findAll('p')[1].get_text().strip()
            else:
                company_name = 'None'
            if len(exp_section.find('div').find('a').findAll('h4'))>1:
                experience=exp_section.find('div').find('a').findAll('h4')[1].find_all('span')[1].get_text().strip()
            else:
                experience = 'None'
            henkilo_job.append([name, job_title, company_name, work_period, experience])



    print(henkilo_info)
    print(henkilo_job)
    print('----------------')
```

```python
df1=pd.DataFrame(henkilo_info, columns=['name','loc','profile_title'])
df1['name']=df1.name.apply(lambda x: x.strip())
df1
```

```python
df2=pd.DataFrame(henkilo_job, columns=['name','job_title','company_name','work_period','experience'])
df2['job_title']=df2.job_title.apply(lambda x : x.replace('\n',' '))
df2['name']=df2.name.apply(lambda x: x.strip())
df2
```

```python
df1.merge(df2, how='left', on='name')
```
